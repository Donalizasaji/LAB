(c) Compute the Discrete-Time Fourier Transform (DTFT) and plot the con tinuous frequency spectrum.


```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
f = 5          # Frequency of the sine wave (Hz)
fs = 1000      # Sampling frequency (Hz)
t = np.arange(0, 1, 1/fs)  # Time vector (1 second duration)

# Generate sine wave
x = np.sin(2 * np.pi * f * t)

# Compute DTFT by evaluating at a fine grid of frequency points
frequencies = np.linspace(-np.pi, np.pi, 1000)  # Frequency vector for continuous spectrum
X_dtft = np.zeros(len(frequencies), dtype=complex)

# DTFT computation
for k in range(len(frequencies)):
    X_dtft[k] = np.sum(x * np.exp(-1j * frequencies[k] * np.arange(len(x))))

# Convert frequency axis to Hz for plotting
frequencies_hz = frequencies * fs / (2 * np.pi)

# Plot the magnitude of the DTFT (continuous frequency spectrum)
plt.figure(figsize=(10, 4))
plt.plot(frequencies_hz, np.abs(X_dtft))
plt.title('DTFT - Continuous Frequency Spectrum')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)
plt.xlim(-50, 50)  # Zoom in to the most relevant part of the spectrum
plt.show()
```

(d) Compute the Discrete Fourier Transform (DFT) and plot the discrete frequency spectrum.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
f = 5          # Frequency of the sine wave (Hz)
fs = 1000      # Sampling frequency (Hz)
N = 1000       # Number of samples
t = np.arange(0, 1, 1/fs)  # Time vector (1 second duration)

# Generate sine wave
x = np.sin(2 * np.pi * f * t)

# Compute DFT using FFT (Fast Fourier Transform)
X_dft = np.fft.fft(x)
frequencies = np.fft.fftfreq(N, 1/fs)  # Frequency vector in Hz

# Plot the magnitude of the DFT (discrete frequency spectrum)
plt.figure(figsize=(10, 4))
plt.stem(frequencies[:N//2], np.abs(X_dft[:N//2]), use_line_collection=True)
plt.title('DFT - Discrete Frequency Spectrum')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)
plt.xlim(0, 50)  # Show relevant frequency range
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the sine waves (5 Hz and 15 Hz)
f1 = 5           # Frequency of first sine wave (Hz)
A1 = 1           # Amplitude of first sine wave
f2 = 15          # Frequency of second sine wave (Hz)
A2 = 0.5         # Amplitude of second sine wave
fs = 1000        # Sampling frequency (Hz)
t = np.arange(0, 1, 1/fs)  # Time vector (1 second duration)

# Generate the two sine waves and composite signal
x1 = A1 * np.sin(2 * np.pi * f1 * t)
x2 = A2 * np.sin(2 * np.pi * f2 * t)
composite_signal = x1 + x2

# Compute DTFT by evaluating the Fourier Transform over a range of frequencies
N = 1000  # Number of frequency points for the DTFT
frequencies = np.linspace(-fs/2, fs/2, N)  # Frequency vector from -fs/2 to fs/2
X_dtft = np.zeros(N, dtype=complex)  # Initialize DTFT result

# Compute the DTFT as a summation over time
for k, f in enumerate(frequencies):
    X_dtft[k] = np.sum(composite_signal * np.exp(-1j * 2 * np.pi * f * t / fs))

# Plot the magnitude of the DTFT (continuous frequency spectrum)
plt.figure(figsize=(10, 4))
plt.plot(frequencies, np.abs(X_dtft))
plt.title('DTFT - Continuous Frequency Spectrum of Composite Signal')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)
plt.xlim(0, 50)  # Limit to the relevant frequency range (0 to 50 Hz)
plt.show()
```

### Inference on Frequency Domain Analysis of Various Signals Using DTFT and DFT

#### Introduction:
Frequency domain analysis using the Discrete-Time Fourier Transform (DTFT) and Discrete Fourier Transform (DFT) is crucial for understanding the characteristics of signals in fields like signal processing and communications.

#### Key Findings:

1. **Signal Decomposition:**
   - DTFT and DFT decompose time-domain signals into frequency components, revealing how different frequencies contribute to the signal's overall behavior.

2. **Frequency Spectrum Characteristics:**
   - **Sinusoidal Signals:** Both transforms show distinct peaks at the frequency of the sinusoid, indicating its presence and amplitude.
   - **Composite Signals:** They reveal multiple frequency peaks corresponding to the constituent sinusoids, demonstrating superposition.
   - **Exponentially Decaying Signals:** The frequency spectrum highlights a concentration of energy at lower frequencies, consistent with the smooth decay observed in the time domain.

3. **Magnitude vs. Phase Information:**
   - The magnitude spectrum indicates the strength of frequency components, while phase information offers insights into the temporal structure of the signal.

4. **Computational Efficiency:**
   - The DFT, particularly with the Fast Fourier Transform (FFT), allows efficient analysis of large datasets, essential for real-time applications.

5. **Limitations and Considerations:**
   - The DFT can introduce spectral leakage, which can be mitigated with windowing techniques. The DTFT provides a continuous view but requires careful sampling to avoid aliasing.

#### Conclusion:
Frequency domain analysis with DTFT and DFT enhances our understanding of signal composition, enabling effective manipulation and interpretation across various applications. These transforms are essential for analyzing complex signals, reinforcing their importance in advancing technology and science.
